## 1. Guitar notes & scales

A scale in music is constructed from a series of tones. The amount varies, major and minor scales, for example, contains seven tones of twelve possible semi-tones. Such series of tones will sound great played together and are often used as foundation for composing or improvisation. They can be played in any order â€“ playing them root to root is only when practicing.

The most common modes in music are **Ionian, Dorian, Phrygian, Lydian, Mixolydian, Aeolian and Locrian**. These goes back to the Ancient Greece and are still used in music today

### 1.2 Some Python functions

```python
def get_notes_key_internal(key, interval):
    """
    Get notes from a key and interval
    """
    start = all_notes.index(key)
    octave = all_notes[start:start + 12]
    result = [octave[idx] for idx in interval]

    return result
```

```python
def get_notes(note, scale):
    """
    Get notes from a scale
    """
    print("Notes of", note, "for scale", scale, "=",
          get_notes_key_internal(note, scales[scale]))
```

```python
def finding_notes(scale):
    """
    Finding guitar notes from a scale
    """
    notes_strings = {idx: 0 for idx in "EADGBE"}

    for key in strings.keys():
        indexes = []

        for note in scale:
            idx = strings[key].index(note)
            indexes.append(idx)

            if idx <= 7:
                indexes.append(idx + 12)

        notes_strings[key] = indexes

    return notes_strings
```

```python
def guitar_scale_view(notemode, mode):
    """
    View notes on a guitar fretboard
    """
    scalemode = scales[mode]
    scale = get_notes_key_internal(notemode, scalemode)

    fig, ax = plt.subplots(figsize=(18, 5))

    for idx in range(1, 7):
        ax.plot([idx for a in range(23)])

    ax.set_axisbelow(True)
    ax.set_xlim([0, 21])
    ax.set_ylim([0.5, 6.5])
    ax.set_facecolor("#654321")  # Brown color
    guitar_plot = finding_notes(scale)

    for y, key in zip([1, 2, 3, 4, 5, 6],
                      'EADGBE'):  # common 6 strings guitar tuning
        for idx in guitar_plot[key]:
            x = idx + 0.5
            p = mpatches.Circle((x, y), 0.3)
            ax.add_patch(p)
            note = strings[key][idx]
            ax.annotate(note, (idx + 0.5, y),
                        color='white',
                        weight='bold',
                        fontsize=12,
                        ha='center',
                        va='center')

    plt.yticks(np.arange(1, 7), ['E', 'A', 'D', 'G', 'B', 'E'])
    plt.xticks(np.arange(21) + 0.5, np.arange(0, 21))
    ax.grid(True)
    plt.title(notemode + " " + mode + " scales", size=15)
    plt.xlabel('Guitar fretboard', size=12)
    plt.ylabel('Guitar strings', size=12)
    plt.savefig(images_dir + "/" + notemode + "_" + mode + "_scale.jpg",
                dpi=300)

    plt.show()
```
